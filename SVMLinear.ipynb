```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification

# Generate a toy dataset (binary classification)
from sklearn.datasets import make_classification

# Fix: Ensure total (informative + redundant) <= n_features
X, y = make_classification(n_samples=100, n_features=2, 
                           n_informative=2, n_redundant=0, 
                           n_classes=2, random_state=42)

# Train SVM model
svm_model = SVC(kernel='linear', C=1.0)
svm_model.fit(X, y)

# Get support vectors
support_vectors = svm_model.support_vectors_
support_indices = svm_model.support_

# Print the indices of support vectors
print("Support Vector Indices:", support_indices)
print("Support Vectors:\n", support_vectors)
```

```python
# Plot decision boundary and support vectors
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], 
            color='black', marker='x', s=100, label="Support Vectors")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.title("SVM Support Vectors")
plt.show()
```

```python
print("Number of Support Vectors:", len(support_vectors))
```
